```python

KEY_FILE = "my_key"
KEY_LEN = 256
FLAG_FILE = "my_flag"


def startup(key_location):
	flag = open(FLAG_FILE).read()
	kf = open(KEY_FILE, "rb").read()
 
	# Only wraps the flag with its size
	start = key_location
	stop = key_location + len(flag)
 
	# Truncates the size of the flag
	key = kf[start:stop]
	key_location = stop  # ?
 
	# for each letter in flag, key it creates a list []\
    # it takes the ASCII code of each letter in FLAG and XOR with the binary of each letter in KEY
    
    # WHY can it make the operation in binary???
    # The flag is ALWAYS in the same key spot
    
    # What I have to do is to get the same portion of the key [0:32]
    # HOW? Thats a good question
	result = list(map(lambda p, k: "{:02x}".format(ord(p) ^ k), flag, key))
	print("This is the encrypted flag!\n{}\n".format("".join(result)))

	return key_location

def encrypt(key_location):
     
	print(key_location)
 
	ui = input("What data would you like to encrypt? ").rstrip()
	if len(ui) == 0 or len(ui) > KEY_LEN:
		return -1

	start = key_location
	stop = key_location + len(ui)

	kf = open(KEY_FILE, "rb").read()
 
 
	# The KEY always wraps around.
	# I could encrypt 50000 - 32 characters to be again at KEY[0]
	# Then I could encrypt 32 more characters get the [0:32] chunk of the key??

	if stop >= KEY_LEN:
		stop = stop % KEY_LEN
		key = kf[start:] + kf[:stop]
	else:
		key = kf[start:stop]
	key_location = stop

	result = list(map(lambda p, k: "{:02x}".format(ord(p) ^ k), ui, key))

	print("Here ya go!\n{}\n".format("".join(result)))

	return key_location


print("******************Welcome to our OTP implementation!******************")
c = startup(0)
while c >= 0:
	c = encrypt(c)
```
